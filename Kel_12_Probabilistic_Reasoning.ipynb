##Bayesian Network

Merupakan sebuah model algoritma probabilitas yang menampilkan nilai dari sebuah variabel dan conditional probability nya (kemungkinan suatu variabel mempengaruhi hasil dari variabel lainnya). Bayesian Network juga dapat menampilkan sekumpulan random variabel dalam bentuk tabel distribusi joint probability.

Bayesian Network menampilkan hubungan antar Random Variabel tadi dalam bentuk directed graph, dimana :



*   Setiap node merepresentasikan sebuah random variabel.
*   Tanda panah dari node X ke node Y artinyan X merupakan parent dari Y (Y dependent ke X)
* Setiap node mempunyai probability distribution

     P(X | Parents(X))

###Model Bayesian Network

Menggunakan library pomegranate untuk dapat membantu pembuatan model Bayesian Network.

```python
from pomegranate import *
```

Membuat node untuk merepresentasikan data hujan dan hujan juga merupakan variabel independen.

```python
hujan = Node(DiscreteDistribution({
    "tidak_hujan": 0.7,
    "ringan": 0.2,
    "deras": 0.1
}), name="hujan")
```

Membuat node untuk merepresentasikan data maintenance dan melakukan joint probability dengan data hujan karena hujan merupakan parent dari maintenance.

```python
maintenance = Node(ConditionalProbabilityTable([
    ["tidak_hujan", "ya", 0.4],
    ["tidak_hujan", "tidak", 0.6],
    ["ringan", "ya", 0.2],
    ["ringan", "tidak", 0.8],
    ["deras", "ya", 0.1],
    ["deras", "tidak", 0.9]
], [hujan.distribution]), name="maintenance")
```

Membuat node untuk merepresentasikan data kereta dan melakukan joint probability dengan data hujan dan maintenance karena hujan dan maintenance merupakan parent dari kereta.

```python
kereta = Node(ConditionalProbabilityTable([
    ["tidak_hujan", "ya", "tepat_waktu", 0.8],
    ["tidak_hujan", "ya", "delay", 0.2],
    ["tidak_hujan", "tidak", "tepat_waktu", 0.9],
    ["tidak_hujan", "tidak", "delay", 0.1],
    ["ringan", "ya", "tepat_waktu", 0.6],
    ["ringan", "ya", "delay", 0.4],
    ["ringan", "tidak", "tepat_waktu", 0.7],
    ["ringan", "tidak", "delay", 0.3],
    ["deras", "ya", "tepat_waktu", 0.4],
    ["deras", "ya", "delay", 0.6],
    ["deras", "tidak", "tepat_waktu", 0.5],
    ["deras", "tidak", "delay", 0.5],
], [hujan.distribution, maintenance.distribution]), name="kereta")
```

Membuat node untuk merepresentasikan data janji dan melakukan joint probability dengan data kereta karena kereta merupakan parent dari janji.

```python
janji = Node(ConditionalProbabilityTable([
    ["tepat_waktu", "tepat waktu", 0.9],
    ["tepat_waktu", "telat", 0.1],
    ["delay", "tepat waktu", 0.6],
    ["delay", "telat", 0.4]
], [kereta.distribution]), name="janji")
```

Lalu kita menginisialisasikan model Bayesian Network menggunakan fungsi yang sudah tersedia di library pomegranate dan menambahkan node hujan, maintenance, kereta, dan janji ke model.

```python
model = BayesianNetwork()
model.add_states(hujan, maintenance, kereta, janji)
```

Menambahkan relasi antar node ke model.

```python
model.add_edge(hujan, maintenance)
model.add_edge(hujan, kereta)
model.add_edge(maintenance, kereta)
model.add_edge(kereta, janji)
```

Melakukan finalisasi model dengan fungsi bake().

```python
model.bake()
```

Test Model :

```python
# Soal : Berapa probabilitas jika Hujan = "tidak hujan", Maintenance = "tidak", Kereta = "tepat waktu", dan Janji = "tepat waktu" ?

probability = model.probability([["tidak_hujan", "tidak", "tepat_waktu", "tepat waktu"]])

# Dari perhitungan model didapatkan probabilitas sekitar 0,34 atau 34% kejadian tersebut terjadi.
print(probability)
```

```python
# Soal : Berapa probabilitas jika Hujan = "tidak hujan", Maintenance = "tidak", Kereta = "tepat waktu", dan Janji = "telat" ?

probability = model.probability([["tidak_hujan", "tidak", "tepat_waktu", "telat"]])

# Dari perhitungan model didapatkan probabilitas sekitar 0,0378 atau 3,78% kejadian tersebut terjadi.
print(probability)
```

Contoh Inference :

```python
#Memprediksi semua variabel acak bila diketahui hanya variabel acak Kereta = "delay"

#Line ini berfungsi untuk menyimpan variabel bukti
predictions = model.predict_proba({
    "kereta": "delay"
})

# Print prediksi dari setiap variabel acak/node
for node, prediction in zip(model.states, predictions):
    if isinstance(prediction, str):
        print(f"{node.name}: {prediction}")
        print(" ")
    else:
        print(f"{node.name}")
        for value, probability in prediction.parameters[0].items():
            print(f"    {value}: {probability:.4f}")
        print(" ")
```

##Model Markov Chain

Model yang dapat memprediksi probabilitas suatu variabel acak dalam kurun waktu tertentu. Markov Chain mengambarkan urutan barisan yang mungkin terjadi dimana setiap kejadian yang diprediksi bergantung pada kejadian yang diprediksi sebelumnya.

Contoh :
* Memprediksi perubahan populasi hewan pada waktu yang akan datang.
* Memprediksi cuaca dalam beberapa hari kedepan.

###Contoh :

Terdapat data sebagai berikut.

Dari data tersebut kita dapat membuat model Markov Chain

```python
#Memanggil library pomegranate
from pomegranate import *
```

```python
#Membuat titik awal yang nantinya akan menentukan hasil selanjutnya saat diinputkan ke model markov chain.

start = DiscreteDistribution({
    "cerah": 0.5,
    "hujan": 0.5
})
```

```python
#Membuat model transisi yang akan memproses titik awal tadi dan menentukan hasil selanjutnya dengan menggunakan data yang kita miliki.

transitions = ConditionalProbabilityTable([
    ["cerah", "cerah", 0.8],
    ["cerah", "hujan", 0.2],
    ["hujan", "cerah", 0.3],
    ["hujan", "hujan", 0.7]
], [start])
```

```python
#Menginisialisasi model Markov Chain
model_markov_chain = MarkovChain([start, transitions])
```

```python
#Menampilkan hasil prediksi dari model yang telah kita buat untuk 10 hari kedepan.
print(model_markov_chain.sample(10))
```

##Hidden Markov Model

Merupakan perkembangan dari model Markov Chain dimana kita dapat memprediksi urutan kejadian dalam jangka waktu tertentu dari sebuah variabel yang dapat kita observasi yang bergantung pada state tersembunyi atau yang tidak dapat kita ketahui datanya.

Contohnya kita dapat memprediksi cuaca (hidden variable) berdasarkan apakah orang-orang menggunakan payung atau tidak (observable variable).

image.png

###Contoh :

Terdapat tabel distribusi probabilitas sebagai berikut, Cuaca (State) merupakan variabel acak yang tidak diketahui dan Payung (Observation) merupakan variabel acak yang dapat kita observasi atau diketahui.

image.png

Berdasarkan data tersebut kita dapat membuat model yang dapat memprediksi Cuaca berdasarkan pengetahuan kita akan data Payung.

```python
#Memanggil library pomegranate
from pomegranate import *
```

```python
#Membuat model probabilitas untuk state "cerah"
cerah = DiscreteDistribution({
    "payung": 0.2,
    "no payung": 0.8
})
```

```python
#Membuat model probabilitas untuk state "hujan"
hujan = DiscreteDistribution({
    "payung": 0.9,
    "no payung": 0.1
})
```

```python
#Mendaftarkan jenis state apa yang mungkin terjadi
states = [cerah, hujan]
```

```python
#Membuat model transisi untuk memprediksi state selanjutnya berdasarkan hasil observasi sekarang
transitions = numpy.array(
    [[0.8, 0.2], #Prediksi besok jika sekarang = "cerah"
     [0.3, 0.7]] #Prediksi besok jika sekarang = "hujan"
)
```

```python
#Membuat model probabilitas sebagai state awal dari model HMM
starts = numpy.array([0.5, 0.5])
```

```python
#Inisialisasi model Hidden Markov
model_hmm = HiddenMarkovModel.from_matrix(
    transitions, states, starts,
    state_names=["cerah", "hujan"]
)

#Finalisasi model menggunakan fungsi bake()
model_hmm.bake()
```

```python
#Contoh data hasil observasi
observations = [
    "payung",
    "payung",
    "no payung",
    "payung",
    "payung",
    "payung",
    "payung",
    "no payung",
    "no payung"
]
```

```python
#Menggunakan fungsi predict() untuk melakukan proses prediksi berdasarkan variabel observations untuk model_hmm
predictions = model_hmm.predict(observations)

#Menampilkan hasil prediksi
for prediction in predictions:
    print(model_hmm.states[prediction].name)
```
